```python
import pandas
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
path = "forEx.csv"
df = pandas.read_table(path, sep=";")
df.head()
```

```python
df = df.drop(columns=["institution"])
df = pandas.get_dummies(df, columns=["country"])
```

```python
y = df['world_rank']
df.drop(columns=['world_rank'], inplace=True)
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = MinMaxScaler().fit_transform(df[numeric_cols])
X = df
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
regressor = KNeighborsRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
score_r2 = r2_score(y_test, y_pred)
rmse_score = root_mean_squared_error(y_test, y_pred)
mae_score = mean_absolute_error(y_test, y_pred)
mape_score = mean_absolute_percentage_error(y_test, y_pred)
print(f"score_r2 = {score_r2}")
print(f"rmse score = {rmse_score}")
print(f"mae score = {mae_score}")
print(f"mape score = {mape_score  * 100}")
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
score_r2 = r2_score(y_test, y_pred)
rmse_score = root_mean_squared_error(y_test, y_pred)
mae_score = mean_absolute_error(y_test, y_pred)
mape_score = mean_absolute_percentage_error(y_test, y_pred)
print(f"score_r2 = {score_r2}")
print(f"rmse score = {rmse_score}")
print(f"mae score = {mae_score}")
print(f"mape score = {mape_score  * 100}")
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
score_r2 = r2_score(y_test, y_pred)
rmse_score = root_mean_squared_error(y_test, y_pred)
mae_score = mean_absolute_error(y_test, y_pred)
mape_score = mean_absolute_percentage_error(y_test, y_pred)
print(f"score_r2 = {score_r2}")
print(f"rmse score = {rmse_score}")
print(f"mae score = {mae_score}")
print(f"mape score = {mape_score  * 100}")
```
