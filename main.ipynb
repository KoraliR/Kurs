```python
import pandas as pd
import glob
import os

df = pd.read_table(r"C:\Users\User\Desktop\Учеба\Практики\Семак 2\ОМО\Курсач\true_output.csv", sep=",")
df = df.drop_duplicates()

```

Посмотрим характеристики

```python
df.info()
df.head()
```

Удалим ненужные столбцы: author, author_type, house_number, residental_complex, street,url

```python
df.drop(columns=['author', 'author_type', 'house_number', 'residential_complex', 'street','url'], inplace=True)
df.head()
```

```python
print(df['accommodation_type'].value_counts())
df['deal_type'].value_counts()
df['location'].value_counts()
```

Удалим столбцы accommodation-type, location и deal_type т.к в них лишь одно уникальное значение. Также удалим столбец underground

```python
df.drop(columns=["accommodation_type", "deal_type", "underground", "location"], inplace=True)
```

Посмотрим, в скольких строках данных содержаться пропуски и сравним с общим размером записей


```python
print(df.shape)
df.isna().sum(axis=0)
```

Удалим строки с пропусками данных в колонке district

```python
df.dropna()
```

Вместо районов будем использовать округа

```python
import json
with open("county_district.json", 'r', encoding='utf-8') as file:
    districts = json.load(file)
new_districts = dict()
for i in districts:
    name = i['name']
    county_id = i['county_id']
    new_districts[name] = county_id
with open("county.json", 'r', encoding='utf-8') as file:
    countis = json.load(file)
new_countis = dict()
for i in countis:
    short_name = i['short']
    id_c = i['id']
    new_countis[id_c] = short_name
print(new_countis)
print(new_districts)

df['district'] = df['district'].replace(new_districts)
df.head()
```

Проверим корректность замены

```python
df['district'].value_counts()
```

Исправим недостатки

```python
countis = {
    "Марьина роща": 3,
    "Филевсий парк": 4,
    "Соколиная гора": 6,
    "Нагатинский затон":8,
    "Филевский парк": 4

}
df['district'] = df['district'].replace(countis)
df['district'].value_counts()
```

Удалим мусор

```python
df['district'] = pd.to_numeric(df['district'], errors='coerce')
df = df.dropna(subset=['district'])
df['district'].value_counts()
```

Применим OneHotEncoder для столбца district

```python
df = pd.get_dummies(df, columns=['district'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

y = df['price']
df.drop(columns=['price'], inplace=True)
X = df
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)



regressor = LinearRegression()
regressor.fit(X_train, y_train)




mine = [10, 18, 3, 60, False, False, False, False, False, False, True, False, False, False]


# or
mine = [mine]

a = regressor.predict(mine)
print(a)

y_pred = regressor.predict(X_test)


score_r2 = r2_score(y_test, y_pred)
mse_score = mean_squared_error(y_test, y_pred)

print(f"score_r2 = {score_r2}")
print(f"mse score = {mse_score}")
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
print(f"MAPE: {mape:.1f}%")

```
